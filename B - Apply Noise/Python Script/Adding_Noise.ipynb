# ADDING NOISE

```python
import os
import cv2
import numpy as np

# ===== Paths =====
# images_dir = r"D:\E - Thesis Collection\DeepFrame-Reconstruction\B - Medial Image Dataset\Brain MRI Images Dataset\Original MRI-Image"
# output_dir = r"D:\E - Thesis Collection\DeepFrame-Reconstruction\B - Medial Image Dataset\Brain MRI Images Dataset\Noise Image"
images_dir = r"D:\E - Thesis Collection\DeepFrame-Reconstruction\B - Medial Image Dataset\Chest CT-Scan images Dataset\Original CT-Scan Image"
output_dir = r"D:\E - Thesis Collection\DeepFrame-Reconstruction\B - Medial Image Dataset\Chest CT-Scan images Dataset\Noise Image"
os.makedirs(output_dir, exist_ok=True)


# ===== Optional Blur =====
def blur_image(image, ksize=(15, 15)):
    """Apply Gaussian blur to an image."""
    return cv2.GaussianBlur(image, ksize, 0)


# ===== Salt & Pepper Noise (5%) =====
def salt_pepper_noise(image, amount=0.05):
    """
    Adds salt & pepper noise to the image.
    amount = fraction of pixels to alter (e.g., 0.05 = 5%)
    """
    noisy = np.copy(image)
    num_salt = np.ceil(amount * image.size * 0.5)
    num_pepper = np.ceil(amount * image.size * 0.5)

    # Add Salt (white pixels)
    coords = [np.random.randint(0, i - 1, int(num_salt)) for i in image.shape]
    noisy[tuple(coords)] = 1

    # Add Pepper (black pixels)
    coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in image.shape]
    noisy[tuple(coords)] = 0

    return noisy


# ===== Process All Images =====
for filename in os.listdir(images_dir):
    if not filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif')):
        continue

    img_path = os.path.join(images_dir, filename)
    image = cv2.imread(img_path)[..., ::-1] / 255.0  # RGB and normalize

    # Step 1: Optional blur
    blurred = blur_image(image)

    # Step 2: Apply salt & pepper noise (5%)
    noisy = salt_pepper_noise(blurred, amount=0.05)

    # Convert back to uint8
    noisy_uint8 = (noisy * 255).astype(np.uint8)

    # Step 3: Save image
    save_path = os.path.join(output_dir, filename)
    cv2.imwrite(save_path, cv2.cvtColor(noisy_uint8, cv2.COLOR_RGB2BGR))

print(f"✅ 5% Salt & Pepper noise added to all images! Check '{output_dir}' for results.")
```
